```python
# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1- Import the csv file


# Provide the file path of your CSV file
file_path = "/content/data.csv"
# Use the read_csv() function to import the CSV file
data = pd.read_csv(file_path)

```

```python
# 2- Display all records

print(data)
```

```python
# 3- Get the list of all attributes (column).
column_list = data.columns.tolist()

# Display the list of attributes
print(column_list)
```

```python
# 4- Fetch top 10 records

print(data.head(10))
```

```python
# 5- Get all unique records

unique_records = data.drop_duplicates()
print(unique_records)
```

```python
# 6- Count the number of unique countries.

unique_countries_count = data['Country/Region'].nunique()
print(unique_countries_count)
```

```python
# 7- Find out the mean, standard deviation, minimum, maximum.


# Calculate the mean
mean_values = data.mean()

# Calculate the standard deviation
std_values = data.std()

# Calculate the minimum
min_values = data.min()

# Calculate the maximum
max_values = data.max()

# Display the mean, standard deviation, minimum, and maximum values
print("Mean values:")
print(mean_values)
print("\nStandard deviation:")
print(std_values)
print("\nMinimum values:")
print(min_values)
print("\nMaximum values:")
print(max_values)
```

```python
# 8- Confirmed Cases, death, recovered cases List by Country

# Group the data by country to get the sum of confirmed cases, deaths, and recovered cases
grouping = data.groupby('Country/Region')['Confirmed', 'Deaths', 'Recovered'].sum().reset_index()

# Display the grouped data
print(grouping)
```

```python
# 9- Confirmed Cases, death, recovered cases List by date.

# Group the data by date to get the sum of confirmed cases, deaths, and recovered cases
# Note:  date = Last Update
grouping = data.groupby('Last Update')['Confirmed', 'Deaths', 'Recovered'].sum().reset_index()

# Display the grouped data
print(grouping)
```

```python
# 10- Confirmed Cases, death, recovered cases List by date and country.

# Group the data by date and country to get the sum of confirmed cases, deaths, and recovered cases
grouping = data.groupby(['Last Update', 'Country/Region'])['Confirmed', 'Deaths', 'Recovered'].sum().reset_index()

# Display the grouped data
print(grouping)
```

```python
# 11- Display the relationship between total cases and recovered cases using scatter plot.


# Create a scatter plot
plt.scatter(data['Confirmed'], data['Recovered'])

# Set labels and title for the plot
plt.xlabel('Total Cases')
plt.ylabel('Recovered Cases')
plt.title('Relationship between Total Cases and Recovered Cases')

# Display the scatter plot
plt.show()

```

```python
# 12- Display the relationship between confirmed, recovered and death cases using scatter plot.


# Create a scatter plot
plt.scatter(data['Confirmed'], data['Recovered'], label='Recovered', color='green')
plt.scatter(data['Confirmed'], data['Deaths'], label='Deaths', color='red')

# Set labels and title for the plot
plt.xlabel('Confirmed Cases')
plt.ylabel('Number of Cases')
plt.title('Relationship between Confirmed, Recovered, and Death Cases')

# Add legend
plt.legend()

# Display the scatter plot
plt.show()
```

```python
# 13- Display the relationship between total cases and confirmed cases of all over the world using scatter plot

# Create a scatter plot
plt.scatter(data['Confirmed'], data['Recovered'])

# Set labels and title for the plot
plt.xlabel('Confirmed Cases')
plt.ylabel('Recovered Cases')
plt.title('Relationship between Total Cases and Confirmed Cases Worldwide')

# Display the scatter plot
plt.show()
```
